```python
# Load the CSV file into a DataFrame
df = pd.read_csv('students.csv')

# Display the DataFrame to verify the data
print(df)
```

```python
passing_students = df.loc[df['Grade'] > 70]

# Display the filtered DataFrame
print(passing_students)
```

```python
passing_students.to_csv('passing_students.csv', index=False)
print("Filtered data exported to passing_students.csv successfully.")
```

```python
df = pd.read_csv('students.csv')
# Display the DataFrame to verify the data
print(df)
```

```python
new_student = {'Name': 'Ahmed Khan', 'Grade': 85, 'Age': 20, 'Percentage': 88.5}

new_student_df = pd.DataFrame([new_student])

# Display the new student's DataFrame to verify
print(new_student_df)
```

```python
df = pd.concat([df, new_student_df], ignore_index=True)

# Display the updated DataFrame to verify the new row has been added
print(df)
```

```python
# Export the updated DataFrame back to the CSV file
df.to_csv('students.csv', mode='a', header=False, index=False)

print("New student data has been successfully appended to students.csv.")
```
